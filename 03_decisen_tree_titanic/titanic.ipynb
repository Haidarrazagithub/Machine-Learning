```python
import pandas as pd
df=pd.read_csv("titanic.csv")
```

```python
import math
median_age = math.floor(df['Age'].mean())
df.Age=df.Age.fillna(median_age)
```

```python
target=df['Survived']
set_df=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked','Survived'],axis='columns')
```

```python
from sklearn.preprocessing import LabelEncoder
le_sex=LabelEncoder()
set_df['sex_l']=le_sex.fit_transform(set_df['Sex'])
input_set=set_df.drop('Sex',axis='columns')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input_set,target,test_size=0.1)
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn import tree
tree.plot_tree(model)
```
